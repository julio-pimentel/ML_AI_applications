```python
credcard_clean=creditcard.drop_duplicates().iloc[:,0:30]
```

```python
credcard_clean.info()
```

```python
credcard_clean.shape[1]
```

```python
cm=credcard_clean.corr()
sns.heatmap(cm,square=True)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
n_col=credcard_clean.shape[1]

fig,ax=plt.subplots(n_col,figsize=(15,150))

for i in range(n_col):
    ax[i].set_title(credcard_clean.columns[i])
    ax[i].hist(credcard_clean.iloc[:,i])


#plt.hist(credcard_clean.iloc[:,0])
```
