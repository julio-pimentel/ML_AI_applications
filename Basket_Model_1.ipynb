```python
#load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#libraries required for MBA
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#read dataset
bread_df=pd.read_csv("BreadBasket.csv")
bread_df.head()
```

```python
#shape of the dataset, 21,293 records
bread_df.shape
```

```python
#check non-null values and variable type
bread_df.info()
```

```python
#check null values again
bread_df.isnull().sum()
```

```python
#find unique values of each column
bread_df.nunique()
```

```python
#transform "Date" column to datetime type
bread_df["Date"]=pd.to_datetime(bread_df["Date"])
#create "Month_year" column
bread_df["Month_year"]=pd.to_datetime(bread_df["Date"]).dt.to_period('M')
#transform "Transaction" to str type
bread_df["Transaction"]=bread_df["Transaction"].astype("str")

bread_df.head()
```

```python
#find unique Items
bread_df["Item"].unique()
```

```python
#check top 10 items
#"NONE" item is in the top 10, check further
bread_df["Item"].value_counts().head(10)
```

```python
#check records that contains "NONE" item
bread_df[bread_df["Item"]=="NONE"].head()
```

```python
#drop items that contains none
bread_df=bread_df[~bread_df["Item"].str.contains("NONE",na=False)]
bread_df.shape
```

```python
#check top 10 again
bread_df["Item"].value_counts().head(10)
```

```python
#check top 10 transactions
bread_df["Transaction"].value_counts().head(10)
```

```python
bread_df.groupby(["Transaction","Item"])["Item"].count()
```

```python
bread_bsk=bread_df.groupby(["Transaction","Item"])["Item"].count().unstack().fillna(0)
bread_bsk
```

```python
basket_model = bread_bsk.applymap(lambda x: 1 if x>0 else 0)
basket_model
```

```python
popular_sets = apriori(basket_model, min_support=0.05, use_colnames=True)
popular_sets
```

```python
rules = association_rules(popular_sets,metric='lift',min_threshold=1)
rules.head(10)
```
