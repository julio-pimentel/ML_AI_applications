```python
import cv2
import numpy as np
```

```python
#Open the image

img=cv2.imread('download.jpg',-1)
cv2.imshow('Nature',img)
k=cv2.waitKey(0)
if k==30:
    cv2.destroyAllWindows()
elif k==ord('s'):
    cv2.imwrite('Nature_one.jpg',img)
    cv2.destroyAllWindows()
```

```python
#Write an Image

cv2.imwrite('Nature_write.jpg',img)
```

```python
#apply gray scale bicubi to the image
from matplotlib import pyplot as plt
img=cv2.imread('download.jpg',-1)
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.xticks([]),plt.yticks([])
plt.show()
```
